```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Define the relative path to the CSV file
csv_file_path = '../../results/final_data_publish_eval.csv'

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)
```

## Reproducibility analysis 

Searched for keywords 'installation', 'setup', 'usage', 'docker', 'containerisation', 'versioning', 'data'. 
(not yet done) Other keywords could be added external liks (wiki, package etc)

```python
# Count the number of True and False values in 'reproduce' and 'security' columns
reproduce_counts = df['reproduce'].value_counts()

# Create a pie chart for reproducibility
plt.figure(figsize=(6, 6))
plt.pie(reproduce_counts, labels=['Found', 'Not found'], autopct='%1.1f%%', startangle=140)
plt.title('Reproducibility keywords in Repositories')
plt.show()
```

## Security analysis
Searched for keywords 'security', 'authentication', 'encryption', 'access control', 'audit', 'logging', 'updates', 'patches', 'penetration testing', 'vulnerability testing'.

```python
# Calculate the counts of 'security' values
security_counts = df['security'].value_counts()

# Create a pie chart for security
plt.figure(figsize=(6, 6))
plt.pie(security_counts, labels=['Found', 'Not found'], autopct='%1.1f%%', startangle=140)
plt.title('Security keywords Repositories')
plt.show()
```
